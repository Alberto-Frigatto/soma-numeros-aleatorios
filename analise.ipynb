# Soma de números aleatórios

Este estudo tem por fim demonstrar que a soma de números aleatórios diminui a aleatoriedade.

### Importando as ferramentas necessárias

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('figure', figsize=(16, 6))
```

### Números aleatórios isolados

Aqui podemos ver que números aleatórios isolados tem a mesma chance de serem sorteados (as pequenas diferenças ocorrem ao acaso).

```python
n_isolados = np.random.random(int(1e6))

ax = sns.histplot(n_isolados, bins=200)
ax.set_title('Números aleatórios isolados', fontsize=16)
```

### Soma de 2 números aleatórios

Aqui podemos ver que o histograma revela um triângulo, ou seja, quanto mais próximo o valor é do limite da distribuição menos ele aparece, mas por quê?

Conforme o valor se aproxima dos limites da distribuição, reduzem-se as possibilidades de formá-lo como resultado da soma.

---

Exemplo:

Os números gerados estão entre 0 e 1, então para formar o valor mínimo possível só há uma forma:

$0 + 0$

E para o valor máximo (2) ocorre o mesmo fenômeno, só há uma possibilidade:

$1 + 1$

Mas para o valor 1 por exemplo, que é a metade do limite superior podemos fazer de muitas maneiras:

- $0 + 1$
- $1 + 0$
- $0.37 + 0.63$
- $0.15 + 0.85$
- $0.4787 + 0.5213$
- ...

```python
soma_2_numeros = np.random.random(int(1e6)) + np.random.random(int(1e6))

ax = sns.histplot(soma_2_numeros, bins=200)
ax.set_title('Soma de 2 números aleatórios', fontsize=16)
```

### Soma de 3 números aleatórios

Aqui podemos ver que o histograma revela uma normal.

Conforme aumentamos a quantidade de números somados, reduzem-se as possibilidades de se formar um número perto dos limites da distribuição.

Podemos ver claramente que a normal mostra que os valores estão mais concentrados no meio dela.

```python
soma_3_numeros = np.random.random(int(1e6)) + np.random.random(int(1e6)) + np.random.random(int(1e6))

ax = sns.histplot(soma_3_numeros, bins=200)
ax.set_title('Soma de 3 números aleatórios', fontsize=16)
```

### Soma de 4 números aleatórios

Aqui podemos ver que o histograma revela uma normal mais.

A partir de 4 números somados, a normal começa a se estreitar, revelando ainda mais a concentração de valores próximos ao meio dela.

```python
soma_4_numeros = np.zeros(int(1e6))

for _ in range(4):
    soma_4_numeros += np.random.random(int(1e6))

ax = sns.histplot(soma_4_numeros, bins=200)
ax.set_title('Soma de 4 números aleatórios', fontsize=16)
```

### Soma de 10.000 números aleatórios

A normal continua se estreitando e concentrando os valores, limites $[0, 10.000]$ nunca apareceram, pois a probabilidade de todos os 10.000 números serem 0 ou todos serem 1 é muito pequena.

E ao passo que continuamos a aumentar a quantidade de números somados, essa distribuição tende ao determinismo.

```python
soma_1e4_numeros = np.zeros(int(1e6))

for _ in np.arange(int(1e4)):
    soma_1e4_numeros += np.random.random(int(1e6))

ax = sns.histplot(soma_1e4_numeros, bins=200)
ax.set_title('Soma de 10.000 números aleatórios', fontsize=16)
```

```python
limites = min(soma_1e4_numeros), max(soma_1e4_numeros)

print(f'Os limites da distribuição são [{limites[0]}, {limites[1]}]')
```

> ## Conclusão
>
> - Conforme somamos números aleatórios, o resultado fica mais previsível, ou seja, a aleatoriedade diminui.
